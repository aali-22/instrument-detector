```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load the audio
filename = "../data/filtered_train/guitar_acoustic_000-021-127.wav"
y, sr = librosa.load(filename, sr=16000)

# Compute Mel Spectrogram
S = librosa.feature.melspectrogram(
    y=y,
    sr=sr,
    n_fft=2048,        # Number of samples per FFT window
    hop_length=512,    # Step size between windows
    n_mels=128         # Number of mel bands (rows of the spectrogram)
)

# Convert to Decibel (log scale)
S_dB = librosa.power_to_db(S, ref=np.max)

# Plot for debugging/understanding
plt.figure(figsize=(10, 4))
librosa.display.specshow(
    S_dB, 
    sr=sr, 
    hop_length=512, 
    x_axis='time', 
    y_axis='mel', 
    cmap='magma'
)
plt.colorbar(format="%+2.0f dB")  # Show dB scale in colorbar, e.g., -80 to 0
plt.title("Mel Spectrogram of guitar(dB Scale)")
plt.tight_layout()
plt.show()
```
